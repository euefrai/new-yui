```python
import sys
from pathlib import Path

root = Path('/workspaces/new-yui')
print('ROOT:', root)

# Compile all python files in the repository to catch syntax errors.
errors = []
for path in sorted(root.rglob('*.py')):
    try:
        import py_compile
        py_compile.compile(str(path), doraise=True)
    except Exception as e:
        errors.append((str(path), str(e)))

print('Compiled files:', len(list(root.rglob('*.py'))))
print('Errors found:', len(errors))
for path, err in errors[:20]:
    print(path)
    print(err)

# Attempt a few imports for main entrypoints.
imports = [
    'config.settings',
    'web_server',
    'cli',
    'core.supabase_client',
    'services.ai_service',
]
imported = []
for mod in imports:
    try:
        __import__(mod)
        imported.append((mod, True, None))
    except Exception as e:
        imported.append((mod, False, str(e)))

print('\nImport results:')
for mod, ok, err in imported:
    print(mod, ok)
    if err:
        print('  ', err)
```
